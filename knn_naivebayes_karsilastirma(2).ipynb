# 🤖 KNN vs Naive Bayes — Kapsamlı Karşılaştırma

---

## 📚 1. KNN (K-Nearest Neighbors) Nedir?

**KNN**, bir örneği sınıflandırmak için eğitim verisindeki **en yakın K komşuya** bakan basit ama güçlü bir algoritmadır.

- **Lazy Learning**: Eğitim sırasında hiçbir şey öğrenmez, tüm işlem tahmin aşamasında yapılır.
- **Instance-based**: Her tahmin için tüm eğitim verisi kullanılır.
- **Non-parametric**: Verinin dağılımı hakkında hiçbir varsayımda bulunmaz.

### Nasıl Çalışır?
```
1. Yeni örnek gelir
2. Tüm eğitim örneklerine uzaklık hesapla
3. En yakın K örneği seç
4. Bu K komşunun çoğunluk sınıfını ata
```

### Uzaklık Metrikleri

**Öklid (Euclidean):**
$$d(x, y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}$$

**Manhattan:**
$$d(x, y) = \sum_{i=1}^{n}|x_i - y_i|$$

**Minkowski (genel form):**
$$d(x, y) = \left(\sum_{i=1}^{n}|x_i - y_i|^p\right)^{1/p}$$

### K Değerinin Önemi
| K Değeri | Sonuç |
|----------|-------|
| K çok küçük (K=1) | Overfitting — gürültüye duyarlı |
| K çok büyük | Underfitting — sınırlar bulanıklaşır |
| K = √N | İyi başlangıç noktası |

### Özellikleri
| Özellik | Değer |
|---------|-------|
| Eğitim süresi | **O(1)** — çok hızlı |
| Tahmin süresi | **O(N×D)** — yavaş |
| Bellek | Tüm veriyi saklar |
| Ölçekleme | **Zorunlu** (StandardScaler) |
| Yorumlanabilirlik | Orta |
| Boyut laneti | Yüksek boyutta kötüleşir |

## 📚 2. Naive Bayes Nedir?

**Naive Bayes**, **Bayes Teoremi**'ne dayanan olasılıksal bir sınıflandırıcıdır. "Naive" (saf) adını, tüm özelliklerin birbirinden **bağımsız** olduğu varsayımından alır.

### Bayes Teoremi

$$P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)}$$

- $P(C|X)$: **Posterior** — X verildiğinde C sınıfının olasılığı
- $P(X|C)$: **Likelihood** — C sınıfında X'in görülme olasılığı
- $P(C)$: **Prior** — C sınıfının genel olasılığı
- $P(X)$: **Evidence** — sabit (paydayı normalleştirmek için)

### Bağımsızlık Varsayımı

$$P(X|C) = \prod_{i=1}^{n} P(x_i|C)$$

**Karar:** $\hat{C} = \arg\max_C P(C) \prod_{i=1}^{n} P(x_i|C)$

### Naive Bayes Türleri
| Tür | Kullanım | Dağılım |
|-----|----------|----------|
| **GaussianNB** | Sürekli özellikler | Normal (Gaussian) |
| **MultinomialNB** | Metin/kelime sayıları | Multinomial |
| **BernoulliNB** | İkili özellikler | Bernoulli |
| **ComplementNB** | Dengesiz metin verisi | Complement |

### Gaussian NB Formülü
$$P(x_i|C) = \frac{1}{\sqrt{2\pi\sigma_C^2}} \exp\left(-\frac{(x_i - \mu_C)^2}{2\sigma_C^2}\right)$$

### Özellikleri
| Özellik | Değer |
|---------|-------|
| Eğitim süresi | **O(N×D)** — çok hızlı |
| Tahmin süresi | **O(C×D)** — çok hızlı |
| Bellek | Sadece istatistikler |
| Ölçekleme | Gerekli değil |
| Yorumlanabilirlik | **Yüksek** |
| Eksik veri | Kolayca işler |
| Bağımsızlık varsayımı | Gerçekte nadiren doğru |

---

## 📊 3. KNN vs Naive Bayes — Özet Karşılaştırma

| Kriter | KNN | Naive Bayes |
|--------|-----|-------------|
| Tip | Instance-based | Probabilistic |
| Öğrenme | Lazy (eğitim yok) | Eager (eğitim hızlı) |
| Tahmin hızı | Yavaş (büyük veride) | **Çok hızlı** |
| Ölçekleme | **Zorunlu** | Gerekli değil |
| Varsayım | Yok | Bağımsızlık |
| Gürültü | Duyarlı (K=1) | Orta |
| Çok boyut | Kötüleşir | Dayanıklı |
| Gerçek zamanlı | ❌ | ✅ |
| Metin sınıflandırma | Zayıf | **Mükemmel** |

## 💻 4. Kütüphaneler ve Veri Hazırlığı

````python
# ─── Temel veri işleme kütüphaneleri ───────────────────────────────────────
import numpy as np                       # Sayısal hesaplamalar ve dizi işlemleri
import pandas as pd                      # Tablo (DataFrame) yapısında veri işleme
import matplotlib.pyplot as plt          # Grafik çizimi
import matplotlib.patches as mpatches   # Grafik üzerinde el ile legend oluşturmak için
import seaborn as sns                    # matplotlib üzerine kurulu, daha şık görselleştirme
import warnings
warnings.filterwarnings('ignore')        # Gereksiz uyarı mesajlarını gizle

# ─── Scikit-learn araçları ──────────────────────────────────────────────────
from sklearn.datasets import load_breast_cancer
# 569 örnek, 30 sayısal özellik içeren gerçek tıbbi veri seti

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
# train_test_split : veriyi eğitim / test olarak böler
# cross_val_score  : k-fold çapraz doğrulama skorları
# GridSearchCV     : hiperparametre ızgara araması (bu notebook'ta opsiyonel kullanım)

from sklearn.preprocessing import StandardScaler
# KNN uzaklık tabanlı çalıştığı için ÖZELLİKLERİN AYNI ÖLÇEĞİNDE OLMASI ŞARTTIR.
# StandardScaler: her özelliği (x - ortalama) / std_sapma ile dönüştürür → N(0,1)

from sklearn.neighbors import KNeighborsClassifier
# K en yakın komşu sınıflandırıcı — eğitim sırasında hiçbir şey öğrenmez (lazy learning)

from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
# GaussianNB  : sürekli özellikler için (bu notebook'ta kullanılacak)
# MultinomialNB : metin / kelime sayısı özellikleri için
# BernoulliNB : 0/1 ikili özellikler için

from sklearn.metrics import (
    accuracy_score,         # Doğru tahmin oranı: (TP+TN) / toplam
    classification_report,  # Precision / Recall / F1 sınıf bazlı özet tablo
    confusion_matrix,       # Hangi sınıf hangisiyle karıştırıldı? (2x2 matris)
    roc_auc_score,          # ROC eğrisinin altındaki alan (0.5=rastgele, 1=mükemmel)
    roc_curve,              # Farklı eşiklerde FPR ve TPR değerleri
    f1_score,               # 2*(Precision*Recall)/(Precision+Recall) harmonik ortalama
    precision_score,        # TP / (TP+FP) → "Pozitif dediğimin kaçı gerçekten pozitif?"
    recall_score            # TP / (TP+FN) → "Gerçek pozitiflerin kaçını yakaladım?"
)

from sklearn.decomposition import PCA
# Principal Component Analysis: 30 boyutlu veriyi 2 boyuta indirgeyerek
# karar sınırını görselleştirmek için kullanacağız

import time   # Eğitim ve tahmin sürelerini ölçmek için

# ─── Görsel ayarlar ─────────────────────────────────────────────────────────
plt.rcParams['figure.figsize'] = (12, 5)      # Varsayılan grafik boyutu (genişlik x yükseklik inç)
plt.rcParams['font.size'] = 12                # Varsayılan yazı boyutu
plt.style.use('seaborn-v0_8-whitegrid')       # Beyaz ızgara arka planlı seaborn stili

print('✅ Tüm kütüphaneler başarıyla yüklendi!')
````

## 🗂️ 5. Veri Seti — Breast Cancer

````python
# ─── Veri setini yükle ──────────────────────────────────────────────────────
data = load_breast_cancer()
# data.data   : (569, 30) boyutlu numpy dizisi — her satır bir hasta, her sütun bir ölçüm
# data.target : (569,) boyutlu dizi — 0=Malignant (kötü huylu), 1=Benign (iyi huylu)

# Daha kolay işlemek için Pandas DataFrame'e dönüştür
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='hedef')

print('📋 Veri Seti Bilgileri')
print('='*40)
print(f'Özellik sayısı  : {X.shape[1]}')
print(f'Örnek sayısı    : {X.shape[0]}')
print(f'Sınıf dağılımı:')
print(f'  0 = Malignant : {sum(y==0)}')
print(f'  1 = Benign    : {sum(y==1)}')

# ─── Eğitim / Test bölmesi ──────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # %20 test, %80 eğitim
    random_state=42,    # Sonuçların tekrar üretilebilir olması için sabit tohum
    stratify=y          # Her bölmede sınıf oranı korunur (dengesiz veri için kritik)
)

# ─── KNN için ölçekleme (ZORUNLU!) ──────────────────────────────────────────
# KNN uzaklık hesaplar. Eğer bir özellik 0-1000 aralığında, diğeri 0-1 aralığındaysa,
# büyük aralıklı özellik her zaman baskın olur → yanıltıcı sonuç!
# StandardScaler ile tüm özellikler aynı ölçeğe (ortalama=0, std=1) getirilir.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
# fit_transform : eğitim verisinden ortalama ve std'yi öğren, sonra uygula
X_test_scaled  = scaler.transform(X_test)
# transform : EĞİTİMDEN öğrenilen değerleri test verisine uygula (veri sızıntısını önler)
X_scaled       = scaler.fit_transform(X)
# Tüm veri: PCA görselleştirme için kullanılacak

print(f'\n✅ Veri bölündü: Eğitim={X_train.shape[0]}, Test={X_test.shape[0]}')
print('✅ StandardScaler uygulandı (KNN için zorunlu, NB için gerekmez)')
````

---
## 🔵 6. KNN Modeli

````python
# ─── En iyi K değerini bul ─────────────────────────────────────────────────
# K=1 → çok hassas (overfitting), K=30 → çok kaba (underfitting)
# En iyi K'yı bulmak için 1-30 arasını deneyerek test doğruluğuna bakıyoruz.

k_range = range(1, 31)   # K=1'den K=30'a kadar tara
k_train_acc = []         # Her K için eğitim doğruluğunu sakla
k_test_acc  = []         # Her K için test doğruluğunu sakla

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    # n_neighbors=k : kaç komşuya bakılacak
    # metric='euclidean' : Öklid uzaklığı = sqrt(Σ(xi-yi)²)
    knn.fit(X_train_scaled, y_train)
    k_train_acc.append(accuracy_score(y_train, knn.predict(X_train_scaled)))
    k_test_acc.append(accuracy_score(y_test,  knn.predict(X_test_scaled)))

# Test doğruluğunun en yüksek olduğu K değerini seç
best_k = k_range[np.argmax(k_test_acc)]
# np.argmax: dizideki maksimum değerin indeksini döner
print(f'🎯 En iyi K değeri: {best_k} (Test Accuracy: {max(k_test_acc):.4f})')

# ─── K vs Accuracy grafiği ──────────────────────────────────────────────────
plt.figure(figsize=(12, 5))
plt.plot(k_range, k_train_acc, 'b-o', markersize=5, linewidth=2, label='Eğitim Acc')
plt.plot(k_range, k_test_acc,  'b--s', markersize=5, linewidth=2, label='Test Acc')
# 'b-o'  : mavi renk, düz çizgi, daire marker
# 'b--s' : mavi renk, kesik çizgi, kare marker

plt.axvline(x=best_k, color='orange', linestyle=':', linewidth=2,
            label=f'En iyi K={best_k}')
# axvline: dikey referans çizgisi — en iyi K'yı işaret eder

# Beklenen görünüm:
# K=1'de eğitim=1.0 (tüm noktalar kendine en yakın → mükemmel eğitim ama overfitting)
# K büyüdükçe eğitim azalır, test önce artar sonra underfitting nedeniyle düşer

plt.xlabel('K Değeri', fontsize=12)
plt.ylabel('Doğruluk', fontsize=12)
plt.title('🔵 KNN — K Değeri vs Doğruluk', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.xticks(k_range)
plt.tight_layout()
plt.show()
````

````python
print('🔵 KNN MODELİ EĞİTİLİYOR...')
print('='*45)

# ─── KNN modelini tanımla ────────────────────────────────────────────────────
knn_model = KNeighborsClassifier(
    n_neighbors=best_k,
    # Yukarıda grid search ile bulunan en iyi K değeri

    metric='euclidean',
    # Uzaklık ölçütü: Öklid → sqrt(Σ(xi-yi)²)
    # Alternatifler: 'manhattan' (Σ|xi-yi|), 'minkowski' (genel form), 'cosine'

    weights='uniform',
    # 'uniform'  : tüm komşular eşit oy kullanır
    # 'distance' : yakın komşular daha fazla ağırlık alır (1/uzaklık)

    algorithm='auto'
    # Uzaklık hesabını hangi veri yapısıyla yapacağını otomatik seç:
    # 'ball_tree' veya 'kd_tree' → büyük veride hızlı
    # 'brute'     → kaba kuvvet, küçük veride yeterli
)

# ─── Eğitim süresi ölç ──────────────────────────────────────────────────────
# KNN'de "eğitim" aslında tüm veriyi bellekte saklamaktan ibarettir.
# Gerçek hesap tahmin aşamasında yapılır → bu yüzden eğitim süresi neredeyse sıfır.
start = time.time()
knn_model.fit(X_train_scaled, y_train)   # Sadece veriyi belleğe al
knn_train_time = time.time() - start

# ─── Tahmin süresi ölç ──────────────────────────────────────────────────────
# Tahmin aşamasında her test örneği için TÜM eğitim verisiyle uzaklık hesaplanır.
# Bu yüzden tahmin süresi eğitim süresinden çok daha uzundur.
start = time.time()
knn_pred = knn_model.predict(X_test_scaled)
# Her test örneği için K en yakın komşuyu bul, çoğunluk oyunu al
knn_pred_time = time.time() - start

knn_prob = knn_model.predict_proba(X_test_scaled)[:, 1]
# predict_proba: K komşu içindeki her sınıfın oranını olasılık olarak döner
# Örnek: 7 komşudan 5'i Benign → P(Benign) = 5/7 ≈ 0.71
# [:, 1] → sadece pozitif sınıf (Benign=1) olasılığı → AUC-ROC için gerekli

# ─── Performans metrikleri ───────────────────────────────────────────────────
knn_acc    = accuracy_score(y_test, knn_pred)
knn_auc    = roc_auc_score(y_test, knn_prob)
knn_f1     = f1_score(y_test, knn_pred)
knn_prec   = precision_score(y_test, knn_pred)
knn_recall = recall_score(y_test, knn_pred)

print(f'✅ Eğitim Süresi   : {knn_train_time:.6f} sn (lazy — çok hızlı!)')
print(f'✅ Tahmin Süresi   : {knn_pred_time:.4f} sn')
print(f'📊 Doğruluk (Acc)  : {knn_acc:.4f}  ({knn_acc*100:.2f}%)')
print(f'📊 AUC-ROC         : {knn_auc:.4f}')
print(f'📊 F1 Skoru        : {knn_f1:.4f}')
print(f'📊 Precision       : {knn_prec:.4f}')
print(f'📊 Recall          : {knn_recall:.4f}')
print()
print('📋 Detaylı Rapor:')
print(classification_report(y_test, knn_pred, target_names=['Malignant', 'Benign']))
````

````python
# ─── KNN Karar Sınırını 2D'de Görselleştir ─────────────────────────────────
# 30 özellikli veriyi 2D'de çizmek imkansız → PCA ile 2 boyuta indir.
# Bu sadece görselleştirme amaçlıdır; asıl model 30 boyutlu uzayda çalışır.

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
# PCA: varyansı en çok açıklayan 2 ekseni (PC1, PC2) bulur.
# explained_variance_ratio_: bu 2 eksenin toplam varyansın ne kadarını açıkladığını verir.

# 2D PCA uzayında train/test bölmesi yap (orijinal bölmeyle aynı random_state)
X_tr_pca, X_te_pca, y_tr_pca, y_te_pca = train_test_split(
    X_pca, y, test_size=0.2, random_state=42, stratify=y
)

# 2D veride KNN eğit (sadece görselleştirme için)
knn_2d = KNeighborsClassifier(n_neighbors=best_k)
knn_2d.fit(X_tr_pca, y_tr_pca)

# ─── Karar sınırı için arka plan ızgara oluştur ──────────────────────────────
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)
# meshgrid: 200x200 = 40.000 noktalı ızgara oluşturur
# Her ızgara noktası için model tahmin yapacak → renkli arka plan elde edilir

Z = knn_2d.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
# xx.ravel()  : 2D ızgara → 1D düz dizi (40.000 x koordinatı)
# np.c_[...]  : iki diziyi sütun olarak birleştir → (40000, 2) dizi
# .reshape()  : tahminleri tekrar 200x200 matrise geri dönüştür

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.25, cmap='coolwarm')
# contourf: doldurulmuş kontur grafiği → karar bölgelerini renklendirir
plt.contour(xx, yy, Z, colors='gray', linewidths=0.5, alpha=0.5)
# contour : sadece sınır çizgisini çizer (doldurmadan)

scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm',
                      edgecolors='white', linewidth=0.5, s=50, alpha=0.85)
# Gerçek veri noktaları: renk = gerçek sınıf
# Arka plan rengi = modelin tahmin ettiği sınıf
# Yanlış sınıflandırılan noktalar: nokta rengi ≠ arka plan rengi

handles = [
    mpatches.Patch(color='#3498db', label='Malignant'),
    mpatches.Patch(color='#e74c3c', label='Benign')
]
plt.legend(handles=handles, fontsize=11)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% varyans)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% varyans)')
plt.title(f'🔵 KNN Karar Sınırı (K={best_k}, PCA ile 2D)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## 🟠 7. Naive Bayes Modeli (Gaussian)

````python
print('🟠 NAIVE BAYES (Gaussian) MODELİ EĞİTİLİYOR...')
print('='*50)

# ─── GaussianNB modelini tanımla ────────────────────────────────────────────
nb_model = GaussianNB(
    var_smoothing=1e-9
    # Varyansın sıfır olduğu durumlarda bölme hatasını önler.
    # Her özelliğin varyansına bu küçük sabit eklenir: var = var + var_smoothing * max(var)
    # Varsayılan değer 1e-9 çoğu durumda yeterlidir.
)

# ─── Eğitim ─────────────────────────────────────────────────────────────────
# GaussianNB eğitimde şunları öğrenir:
#   - Her sınıfın önsel olasılığı: P(C) = o sınıftaki örnek sayısı / toplam
#   - Her özellik için her sınıftaki ortalama (θ) ve varyans (σ²)
#   → Bu değerler nb_model.theta_ ve nb_model.var_ içinde saklanır
# NOT: KNN'den farklı olarak ölçekleme GEREKMEZ.
# Çünkü NB her özelliği bağımsız değerlendirir ve kendi dağılımını öğrenir.
start = time.time()
nb_model.fit(X_train, y_train)
nb_train_time = time.time() - start

# ─── Tahmin ─────────────────────────────────────────────────────────────────
start = time.time()
nb_pred = nb_model.predict(X_test)
# Tahmin: her sınıf için P(C) × Π P(xi|C) hesapla, en büyüğünü seç
nb_pred_time = time.time() - start

nb_prob = nb_model.predict_proba(X_test)[:, 1]
# predict_proba: sınıflar arası normalize edilmiş posterior olasılıkları
# [:, 1] → Benign (pozitif sınıf) olasılığı — AUC hesabı için

# ─── Performans metrikleri ───────────────────────────────────────────────────
nb_acc    = accuracy_score(y_test, nb_pred)
nb_auc    = roc_auc_score(y_test, nb_prob)
nb_f1     = f1_score(y_test, nb_pred)
nb_prec   = precision_score(y_test, nb_pred)
nb_recall = recall_score(y_test, nb_pred)

print(f'✅ Eğitim Süresi   : {nb_train_time:.6f} sn (ultra hızlı!)')
print(f'✅ Tahmin Süresi   : {nb_pred_time:.6f} sn (ultra hızlı!)')
print(f'📊 Doğruluk (Acc)  : {nb_acc:.4f}  ({nb_acc*100:.2f}%)')
print(f'📊 AUC-ROC         : {nb_auc:.4f}')
print(f'📊 F1 Skoru        : {nb_f1:.4f}')
print(f'📊 Precision       : {nb_prec:.4f}')
print(f'📊 Recall          : {nb_recall:.4f}')
print()
print('📋 Detaylı Rapor:')
print(classification_report(y_test, nb_pred, target_names=['Malignant', 'Benign']))
````

````python
# ─── Naive Bayes'in Öğrendiği Gaussian Dağılımları Görselleştir ────────────
# NB eğitimde her özellik için her sınıfta ortalama ve varyans öğrenir.
# Bu hücre o öğrenilen dağılımları çiziyor → modelin içini görüyoruz!
# İki sınıfın dağılımları ne kadar az örtüşürse özellik o kadar ayırt edicidir.

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()   # 2x2 grid → 4 elemanlı 1D diziye çevir (döngüde kolayca kullanmak için)

colors = ['#e74c3c', '#2ecc71']
class_labels = ['Malignant', 'Benign']

for i in range(4):    # İlk 4 özellik için
    feat = data.feature_names[i]
    for c in range(2):    # Her sınıf için (0=Malignant, 1=Benign)
        mu    = nb_model.theta_[c, i]
        # theta_[c, i]: c. sınıfın i. özellik için öğrenilen ortalaması (μ)

        sigma = np.sqrt(nb_model.var_[c, i])
        # var_[c, i]: c. sınıfın i. özellik için öğrenilen varyansı (σ²)
        # np.sqrt → standart sapma (σ) alıyoruz

        x_range = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
        # μ ± 4σ aralığı: Normal dağılımın %99.99'unu kapsar

        pdf = (1 / (sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x_range - mu) / sigma)**2)
        # Gaussian PDF formülü: P(x|C) = (1/σ√2π) × exp(-(x-μ)²/2σ²)
        # Bu, NB'nin tahmin aşamasında kullandığı tam formüldür

        axes[i].plot(x_range, pdf, color=colors[c], linewidth=2.5, label=class_labels[c])
        axes[i].fill_between(x_range, pdf, alpha=0.15, color=colors[c])  # Alan gölgesi
        axes[i].axvline(mu, color=colors[c], linestyle='--', alpha=0.6, linewidth=1.5)
        # Dikey kesik çizgi: ortalama μ'yü işaret eder

    axes[i].set_title(f'{feat}', fontsize=11, fontweight='bold')
    axes[i].set_xlabel('Değer')
    axes[i].set_ylabel('Olasılık Yoğunluğu')
    axes[i].legend(fontsize=10)

plt.suptitle('🟠 Naive Bayes — Öğrenilen Gaussian Dağılımlar (İlk 4 Özellik)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

````python
# ─── 5-Fold Cross Validation ────────────────────────────────────────────────
# Tek bir train/test bölmesi şansa bağlı olabilir.
# CV: veriyi 5 eşit parçaya böler, her seferinde 1 parça test kalanlar eğitim.
# Sonuç: 5 farklı skor → ortalama ve std ile daha güvenilir performans tahmini.

print('🔄 5-Fold Cross Validation...')
print('='*45)

# KNN için ölçekleme Pipeline içinde yapılır:
# Pipeline: adımları zincirleme birleştirir ve her fold'da
# SADECE EĞİTİM verisiyle fit eder → veri sızıntısı olmaz
from sklearn.pipeline import Pipeline
knn_pipe = Pipeline([
    ('scaler', StandardScaler()),                           # Adım 1: ölçekle
    ('knn',    KNeighborsClassifier(n_neighbors=best_k))   # Adım 2: KNN
])
# Pipeline olmadan: tüm veriyi ölçekleyip sonra CV yapsaydık
# test verisinin bilgisi eğitime sızmış olurdu (data leakage)

knn_cv = cross_val_score(knn_pipe, X, y, cv=5, scoring='accuracy')
# X: ölçeklenmemiş ham veri — Pipeline içinde otomatik ölçeklenir
nb_cv  = cross_val_score(nb_model, X, y, cv=5, scoring='accuracy')
# NB ölçekleme gerektirmez, doğrudan ham veriyle

print(f'🔵 KNN          CV: {knn_cv.round(4)} → Ort: {knn_cv.mean():.4f} ± {knn_cv.std():.4f}')
print(f'🟠 Naive Bayes  CV: {nb_cv.round(4)}  → Ort: {nb_cv.mean():.4f} ± {nb_cv.std():.4f}')
# Küçük std → model farklı veri bölmelerinde tutarlı (kararlı)
# Büyük std → model bazı bölmelerde iyi, bazılarında kötü (kararsız)

# ─── CV görselleştir ────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))
x_fold = np.arange(5)   # [0, 1, 2, 3, 4] → 5 fold için x koordinatları

ax.plot(x_fold, knn_cv, 'b-o', markersize=8, linewidth=2,
        label=f'KNN (ort={knn_cv.mean():.4f})')
ax.plot(x_fold, nb_cv, '-s', color='orange', markersize=8, linewidth=2,
        label=f'Naive Bayes (ort={nb_cv.mean():.4f})')

# Gölgeli bant: ortalama ± 1 standart sapma aralığı
# Dar bant → kararlı model; geniş bant → fold'a göre değişken
ax.fill_between(x_fold,
                knn_cv.mean() - knn_cv.std(),
                knn_cv.mean() + knn_cv.std(),
                alpha=0.1, color='blue')
ax.fill_between(x_fold,
                nb_cv.mean() - nb_cv.std(),
                nb_cv.mean() + nb_cv.std(),
                alpha=0.1, color='orange')

ax.set_xticks(x_fold)
ax.set_xticklabels([f'Fold {i+1}' for i in range(5)])
ax.set_ylabel('Accuracy', fontsize=12)
ax.set_title('5-Fold Cross Validation', fontsize=14, fontweight='bold')
ax.legend(fontsize=12)
plt.tight_layout()
plt.show()
````

````python
# ─── Farklı Veri Boyutlarında Eğitim & Tahmin Süresi ───────────────────────
# KNN ve NB'nin hız farkı veri büyüdükçe nasıl değişiyor?
# KNN: eğitim O(1) ama tahmin O(N×D) → N büyüdükçe tahmin yavaşlar
# NB : eğitim O(N×D) ama tahmin O(C×D) → çok hızlı, N'den bağımsız

sizes = [50, 100, 200, 300, 400, 455]   # Farklı eğitim boyutları (maksimum 455 = %80 veri)
knn_train_times, nb_train_times = [], []
knn_pred_times,  nb_pred_times  = [], []

for s in sizes:
    Xs    = X_train_scaled[:s]   # Ölçeklenmiş eğitim verisinin ilk s örneği (KNN için)
    Xs_nb = X_train.iloc[:s]     # Ham eğitim verisi (NB için — ölçekleme gerekmez)
    ys    = y_train.iloc[:s]     # İlgili etiketler

    # ── KNN eğitim süresi ──
    m_knn = KNeighborsClassifier(n_neighbors=best_k)
    t = time.time()
    m_knn.fit(Xs, ys)            # Sadece veriyi belleğe al → çok hızlı
    knn_train_times.append(time.time() - t)

    # ── KNN tahmin süresi ──
    t = time.time()
    m_knn.predict(X_test_scaled) # 114 test örneği × s eğitim örneği → uzaklık hesabı
    knn_pred_times.append(time.time() - t)

    # ── NB eğitim süresi ──
    m_nb = GaussianNB()
    t = time.time()
    m_nb.fit(Xs_nb, ys)          # Ortalama ve varyans hesapla → hızlı ama O(N×D)
    nb_train_times.append(time.time() - t)

    # ── NB tahmin süresi ──
    t = time.time()
    m_nb.predict(X_test)         # Sadece istatistikler kullanılır → N'den bağımsız
    nb_pred_times.append(time.time() - t)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Sol: Eğitim süresi karşılaştırması
axes[0].plot(sizes, knn_train_times, 'b-o', markersize=7, linewidth=2, label='KNN')
axes[0].plot(sizes, nb_train_times,  '-s', color='orange', markersize=7, linewidth=2, label='Naive Bayes')
# Beklenti: KNN eğitimi neredeyse sıfır, NB biraz daha uzun ama her ikisi de çok hızlı
axes[0].set_title('Eğitim Süresi', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Eğitim Örnek Sayısı')
axes[0].set_ylabel('Süre (sn)')
axes[0].legend()

# Sağ: Tahmin süresi karşılaştırması
axes[1].plot(sizes, knn_pred_times, 'b-o', markersize=7, linewidth=2, label='KNN')
axes[1].plot(sizes, nb_pred_times,  '-s', color='orange', markersize=7, linewidth=2, label='Naive Bayes')
# Beklenti: KNN tahmin süresi N ile büyür, NB tahmin süresi sabit kalır
axes[1].set_title('Tahmin Süresi', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Eğitim Örnek Sayısı (tahmin edilen = sabit 114 örnek)')
axes[1].set_ylabel('Süre (sn)')
axes[1].legend()

plt.suptitle('Süre Karşılaştırması: Eğitim vs Tahmin', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
````

## 🎯 10. Hangi Algoritmayı Kullanmalısınız?

### ✅ KNN Kullanın — Eğer:
- Veri seti **küçük-orta** boyuttaysa (büyükte yavaşlar)
- Verinin yapısı **doğrusal değil** ve karmaşık sınır varsa
- **Anomali tespiti** yapıyorsanız
- Özellikler arasındaki **yerel örüntüler** önemliyse
- Hızlı prototip gerekiyorsa ve **ölçeklemeyi** yapabiliyorsanız

### ✅ Naive Bayes Kullanın — Eğer:
- **Metin sınıflandırması** (spam, duygu analizi) yapıyorsanız
- **Gerçek zamanlı** tahmin gerekiyorsa (çok hızlı)
- Veri seti **büyük** ve bellekte saklamak istemiyorsanız
- **Eksik veri** varsa
- Özellikler birbirinden **gerçekten bağımsızsa** (nadir ama mümkün)
- **Yorumlanabilir** olasılık çıktısı istiyorsanız

## 🚀 11. Özet

| | KNN | Naive Bayes |
|---|---|---|
| **En iyi durum** | Küçük, temiz, ölçeklenmiş veri | Büyük veri, metin, gerçek zamanlı |
| **Güçlü yönü** | Varsayımsız, esnek sınır | Hız, basitlik, olasılık çıktısı |
| **Zayıf yönü** | Bellek & tahmin yavaşlığı | Bağımsızlık varsayımı |
| **Ölçekleme** | **Zorunlu** | Gerekli değil |
| **Endüstri kullanımı** | Öneri sistemleri, anomali tespiti | Spam filtresi, NLP, tıbbi teşhis |

> 💡 **Kural:** KNN sezgisel ve esnek, Naive Bayes hızlı ve yorumlanabilir. İkisi de baseline model olarak mükemmel.
