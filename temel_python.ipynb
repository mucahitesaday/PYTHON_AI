## 3️⃣ String (Metin) İşlemleri

```python
# Örnek: String metodları
metin = "Merhaba Python!"

print(metin.upper())       # Büyük harf
print(metin.lower())       # Küçük harf
print(metin.replace("Python", "Dünya"))  # Değiştirme
print(len(metin))          # Uzunluk
print(metin[0:7])          # Dilimleme
```

## 5️⃣ Döngüler (for / while)

## 6️⃣ Listeler

```python
# Örnek
meyveler = ["elma", "armut", "çilek", "muz"]

print(meyveler[0])          # İlk eleman
print(meyveler[-1])         # Son eleman
print(meyveler[1:3])        # Dilimleme
meyveler.append("kivi")     # Ekleme
meyveler.remove("armut")    # Silme
print(meyveler)
```

## 9️⃣ Hata Yönetimi (try / except)

```python
# Örnek
try:
    sonuc = 10 / 0
except ZeroDivisionError:
    print("Sıfıra bölme hatası!")
except Exception as e:
    print(f"Beklenmeyen hata: {e}")
finally:
    print("İşlem tamamlandı.")
```
